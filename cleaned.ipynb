## Predictive Modeling Section

Step 1: Import machine learning libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.ensemble import HistGradientBoostingRegressor, RandomForestClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import classification_report, accuracy_score, f1_score, ConfusionMatrixDisplay
```

Step 2: Create a modeling copy of the cleaned dataset

```python
df_model = df_clean.copy()

df_model.shape
```

This creates a separate copy for modeling.
The cleaned dataset stays untouched, and df_model becomes the dataset we use for machine learning.

Step 3: Select useful model features

```python
candidate_features = [
    "year",
    "odometer",
    "vehicle_age",
    "lat",
    "long",
    "manufacturer",
    "condition",
    "cylinders",
    "fuel",
    "title_status",
    "transmission",
    "drive",
    "type",
    "state",
    "mileage_risk",
    "age_risk",
    "title_risk"
]

feature_cols = [col for col in candidate_features if col in df_model.columns]

feature_cols
```

This selects the columns that will help predict car price.

Step 4: Prepare the regression dataset

```python
df_regression = df_model[feature_cols + ["price"]].copy()

df_regression = df_regression[df_regression["price"].notnull()]
df_regression = df_regression[df_regression["price"] > 500]
df_regression = df_regression[df_regression["price"] <= 150000]

df_regression.shape
```

This prepares the data for price prediction.
I am trying to predict the price, I also make sure the price values are realistic.

Step 5: Sample the data for faster model training

```python
sample_size = 100000

if len(df_regression) > sample_size:
    df_regression_sample = df_regression.sample(sample_size, random_state=42)
else:
    df_regression_sample = df_regression.copy()

df_regression_sample.shape
```

This creates a sample of the dataset for training.
the dataset is BIG, using a sample helps the model train faster. 100,000 rows are still big enough to learn patterns.

This is a practical modeling decision. Balancing model quality with computer performance.

Step 6: Split the data into training and testing sets

```python
X = df_regression_sample[feature_cols]
y = df_regression_sample["price"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

X_train.shape, X_test.shape
```

This splits the data into two parts.
The training set teaches the model. The testing set checks whether the model learned well on data it has not seen before.
This is important because I do not only want the model to memorize the dataset. I want it to make useful predictions on new car listings.

Step 7: Separate numeric and categorical features

```python
numeric_features = [
    col for col in ["year", "odometer", "vehicle_age", "lat", "long"]
    if col in feature_cols
]

categorical_features = [
    col for col in feature_cols
    if col not in numeric_features
]

numeric_features, categorical_features
```

This separates number columns from text/category columns.
Numbers like year, odometer, and vehicle_age can be used directly after cleaning. Text columns like manufacturer, condition, and title_status need to be encoded so the model can understand them.
Machine learning models do not understand words naturally. We have to turn categories into a numeric format.

Step 8: Build a baseline Ridge regression model 

```python
try:
    one_hot_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=True)
except TypeError:
    one_hot_encoder = OneHotEncoder(handle_unknown="ignore", sparse=True)

linear_preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), numeric_features),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", one_hot_encoder)
        ]), categorical_features)
    ]
)

ridge_model = TransformedTargetRegressor(
    regressor=Pipeline([
        ("preprocess", linear_preprocessor),
        ("model", Ridge(alpha=1.0))
    ]),
    func=np.log1p,
    inverse_func=np.expm1
)

ridge_model.fit(X_train, y_train)
```

Step 9: Build an advanced gradient boosting regression model

```python
tree_preprocessor = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), numeric_features),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
        ]), categorical_features)
    ]
)

boosted_price_model = TransformedTargetRegressor(
    regressor=Pipeline([
        ("preprocess", tree_preprocessor),
        ("model", HistGradientBoostingRegressor(
            max_iter=200,
            learning_rate=0.08,
            max_leaf_nodes=31,
            random_state=42
        ))
    ]),
    func=np.log1p,
    inverse_func=np.expm1
)

boosted_price_model.fit(X_train, y_train)
```

This builds a more advanced model called a gradient boosting regressor.
This model can capture nonlinear patterns. That means it can learn more complex relationships, such as how price changes differently depending on mileage, age, manufacturer, title status, and condition.
This is stronger than only using a simple linear model.

This is an advanced tree-based model. It learns patterns by making many small decision trees that develop step by step.
Ridge is the simple baseline model, and Gradient Boosting is the smarter model.

Both models trained successfully. The visual boxes simply show the steps inside each model pipeline. The Ridge model is useful as a baseline because it is simpler and easier to compare against. The Gradient Boosting model is more advanced because it can capture complex relationships between price, mileage, age, manufacturer, condition, and title risk.

Now the important next step is to evaluate them using MAE, RMSE, and R² to see which model predicts used-car prices better.

Step 10: Evaluate both regression models

```python
def evaluate_regression_model(model_name, model, X_test, y_test):
    predictions = model.predict(X_test)
    predictions = np.maximum(predictions, 0)

    mae = mean_absolute_error(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    r2 = r2_score(y_test, predictions)

    return {
        "model": model_name,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    }

regression_results = pd.DataFrame([
    evaluate_regression_model("Ridge Regression", ridge_model, X_test, y_test),
    evaluate_regression_model("Gradient Boosting Regression", boosted_price_model, X_test, y_test)
])

regression_results.sort_values(by="RMSE")
```

This evaluates both models.

MAE means the average prediction error in dollars.
RMSE also measures error, but it punishes large mistakes more.
R2 shows how much of the price variation the model explains.

A lower MAE and RMSE is better. A higher R2 is better.
This table helps compare the simple model against the advanced model.

here we can see that Gradient boosting Regression performed better than Ridge regression 

Step 11: Choose the better regression model

```python
best_model_name = regression_results.sort_values(by="RMSE").iloc[0]["model"]

if best_model_name == "Gradient Boosting Regression":
    best_price_model = boosted_price_model
else:
    best_price_model = ridge_model

best_model_name
```

This chooses the better model based on RMSE.
This is a good modeling habit because I am not just picking the fancy model automatically. I am letting the evaluation results decide.

Step 12: Plot actual price vs predicted price

```python
y_pred = best_price_model.predict(X_test)
y_pred = np.maximum(y_pred, 0)

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.2, color="#4C78A8")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.tight_layout()
plt.show()
```

the graph shows a diagonal pattern, which means the model is learning a real relationship: as actual car prices increase, predicted prices usually increase too.

However, the model is not perfect. For very expensive cars, especially above around $80,000–$100,000, the model often predicts lower prices than the actual listing price. That means the model is better at predicting normal used-car prices than very high-end or unusual listings.

Step 13: Create expected price and price gap features

```python
df_model["expected_price"] = best_price_model.predict(df_model[feature_cols])
df_model["expected_price"] = np.maximum(df_model["expected_price"], 0)

df_model["price_gap"] = df_model["price"] - df_model["expected_price"]

df_model["price_gap_percent"] = (
    df_model["price_gap"] / df_model["expected_price"].replace(0, np.nan)
) * 100

df_model["price_gap_percent"] = df_model["price_gap_percent"].replace([np.inf, -np.inf], np.nan)
df_model["price_gap_percent"] = df_model["price_gap_percent"].fillna(0)

df_model[["price", "expected_price", "price_gap", "price_gap_percent"]].head()
```

If price_gap is positive, the car is listed above the expected price meaning (a possible overpriced listing)
If price_gap is negative, the car is listed below the expected price.

This creates a machine-learning-based way to identify potentially overpriced or undervalued vehicles.

Step 14: Visualize price gap percent

```python
price_gap_plot = df_model[
    (df_model["price_gap_percent"] >= -100) &
    (df_model["price_gap_percent"] <= 100)
]

price_gap_plot["price_gap_percent"].plot(kind="hist", bins=50, figsize=(10, 5), color="#1F3A5F" )

plt.title("Distribution of Price Gap Percent")
plt.xlabel("Price Gap Percent")
plt.ylabel("Number of Listings")
plt.tight_layout()
plt.show()
```

## Classification Logic for Buyer-Risk Categories

Step 15: Create model-assisted buyer-risk labels

```python
def assign_buyer_risk(row):
    if row["title_risk"] == "very high title risk":
        return "avoid"
    elif row["mileage_risk"] == "very high mileage" and row["price_gap_percent"] > 10:
        return "high-mileage risk"
    elif row["price_gap_percent"] > 25:
        return "overpriced"
    elif row["mileage_risk"] == "very high mileage":
        return "high-mileage risk"
    elif (
        row["price_gap_percent"] <= -15 and
        row["title_risk"] == "low title risk" and
        row["mileage_risk"] in ["low mileage", "moderate mileage"] and
        row["age_risk"] in ["newer used car", "moderately aged"]
    ):
        return "smart buy"
    else:
        return "fair deal"

df_model["buyer_risk_category"] = df_model.apply(assign_buyer_risk, axis=1)

df_model["buyer_risk_category"].value_counts()
```

This creates buyer-risk categories using both machine learning and business logic.

It uses:
expected price prediction
actual price
title risk
mileage risk
age risk

This is stronger than simple rule-based scoring because the model first estimates what the car should cost, then the classification logic uses that expected price to judge whether the listing looks risky.

Machine Learning Classification Model

Step 17: Prepare data for classification

```python
classification_features = [
    "price",
    "expected_price",
    "price_gap_percent",
    "year",
    "odometer",
    "vehicle_age",
    "lat",
    "long",
    "manufacturer",
    "condition",
    "cylinders",
    "fuel",
    "title_status",
    "transmission",
    "drive",
    "type",
    "state",
    "mileage_risk",
    "age_risk",
    "title_risk",
    "vin_is_fake_or_unknown"
]

classification_features = [col for col in classification_features if col in df_model.columns]

df_classification = df_model[classification_features + ["buyer_risk_category"]].copy()

df_classification = df_classification.dropna(subset=["buyer_risk_category"])

df_classification.shape
```

This prepares the classification dataset.
The target is buyer_risk_category.

Step 18: Split classification data into training and testing sets

```python
X_class = df_classification[classification_features]
y_class = df_classification["buyer_risk_category"]

X_class_train, X_class_test, y_class_train, y_class_test = train_test_split(
    X_class,
    y_class,
    test_size=0.20,
    random_state=42,
    stratify=y_class
)

X_class_train.shape, X_class_test.shape
```

This splits the classification data into training and testing sets.
The model learns from the training data and is evaluated on the testing data.
The stratify=y_class part keeps the same class balance in both sets. This matters because some categories may be much more common than others.

Step 19: Build a Random Forest classification model

```python
classification_numeric_features = [
    col for col in [
        "price",
        "expected_price",
        "price_gap_percent",
        "year",
        "odometer",
        "vehicle_age",
        "lat",
        "long"
    ]
    if col in classification_features
]

classification_categorical_features = [
    col for col in classification_features
    if col not in classification_numeric_features
]

classification_preprocessor = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), classification_numeric_features),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
        ]), classification_categorical_features)
    ]
)

risk_classifier = Pipeline([
    ("preprocess", classification_preprocessor),
    ("model", RandomForestClassifier(
        n_estimators=200,
        max_depth=18,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

risk_classifier.fit(X_class_train, y_class_train)
```

This builds a Random Forest classifier.
Random Forest is a great machine learning model as it combines many decision trees. It can capture nonlinear patterns and interactions between features.

Step 20: Evaluate the classification model

```python
class_predictions = risk_classifier.predict(X_class_test)

classification_accuracy = accuracy_score(y_class_test, class_predictions)
classification_f1 = f1_score(y_class_test, class_predictions, average="weighted")

classification_accuracy, classification_f1
```

This gives two important evaluation scores.

Accuracy is the percentage of correct predictions.
Weighted F1-score is good when categories are imbalanced as it considers both precision and recall.
F1-score is especially useful because I care about catching risky listings, not just predicting the most common category.

Step 20: Create classification features without direct leakage

```python
classification_features_no_leakage = [
    "price",
    "expected_price",
    "year",
    "odometer",
    "vehicle_age",
    "lat",
    "long",
    "manufacturer",
    "condition",
    "cylinders",
    "fuel",
    "title_status",
    "transmission",
    "drive",
    "type",
    "state",
    "vin_is_fake_or_unknown"
]

classification_features_no_leakage = [
    col for col in classification_features_no_leakage
    if col in df_model.columns
]

classification_features_no_leakage
```

This creates a new list of features for the classification model.

The important change is that I removed columns like price_gap_percent, mileage_risk, age_risk, and title_risk because those were too directly connected to how buyer_risk_category was created.

Before, the model was accidentally being shown the answer key. Now, we are making the model work harder by using more basic vehicle information, such as price, expected price, year, odometer, manufacturer, condition, title status, and vehicle type.

```python
X_class_clean = df_classification_clean[classification_features_no_leakage]
y_class_clean = df_classification_clean["buyer_risk_category"]

X_class_train_clean, X_class_test_clean, y_class_train_clean, y_class_test_clean = train_test_split(
    X_class_clean,
    y_class_clean,
    test_size=0.20,
    random_state=42,
    stratify=y_class_clean
)

X_class_train_clean.shape, X_class_test_clean.shape
```

This splits the corrected classification data into training and testing sets.
The training set teaches the model.

Step 23: Build the corrected Random Forest classifier

```python
classification_numeric_features_clean = [
    col for col in [
        "price",
        "expected_price",
        "year",
        "odometer",
        "vehicle_age",
        "lat",
        "long"
    ]
    if col in classification_features_no_leakage
]

classification_categorical_features_clean = [
    col for col in classification_features_no_leakage
    if col not in classification_numeric_features_clean
]

classification_preprocessor_clean = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), classification_numeric_features_clean),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
        ]), classification_categorical_features_clean)
    ]
)

risk_classifier_clean = Pipeline([
    ("preprocess", classification_preprocessor_clean),
    ("model", RandomForestClassifier(
        n_estimators=200,
        max_depth=14,
        min_samples_leaf=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

risk_classifier_clean.fit(X_class_train_clean, y_class_train_clean)
```

Random Forest is still a strong model because it combines many decision trees. It can learn patterns using the features to conclude if the lsiting is risky or not

Step 24B: Make the Smart Buy category more conservative

```python
class_probabilities = risk_classifier_clean.predict_proba(X_class_test_clean)

class_labels = risk_classifier_clean.classes_

smart_buy_index = list(class_labels).index("smart buy")

default_predictions = risk_classifier_clean.predict(X_class_test_clean)

adjusted_predictions = default_predictions.copy()

smart_buy_probability = class_probabilities[:, smart_buy_index]

adjusted_predictions[
    (default_predictions == "smart buy") & 
    (smart_buy_probability < 0.75)
] = "fair deal"

adjusted_accuracy = accuracy_score(y_class_test_clean, adjusted_predictions)
adjusted_f1 = f1_score(y_class_test_clean, adjusted_predictions, average="weighted")

adjusted_accuracy, adjusted_f1
```

Part 2: Turn it into a reusable function

Step 41: Create helper functions for feature engineering

```python
def mileage_risk_category(odometer):
    if odometer < 50000:
        return "low mileage"
    elif odometer < 100000:
        return "moderate mileage"
    elif odometer < 150000:
        return "high mileage"
    else:
        return "very high mileage"


def age_risk_category(vehicle_age):
    if vehicle_age <= 5:
        return "newer used car"
    elif vehicle_age <= 10:
        return "moderately aged"
    elif vehicle_age <= 15:
        return "older car"
    else:
        return "very old car"


def title_risk_category(title_status):
    if title_status == "clean":
        return "low title risk"
    elif title_status in ["rebuilt", "salvage"]:
        return "high title risk"
    elif title_status in ["lien", "missing", "parts only"]:
        return "very high title risk"
    else:
        return "unknown title risk"
```

Create the reusable buyer-risk prediction function

```python
def predict_buyer_risk(car_information):
    car_df = pd.DataFrame([car_information])

    car_df["manufacturer"] = car_df["manufacturer"].astype(str).str.strip().str.lower()
    car_df["condition"] = car_df["condition"].astype(str).str.strip().str.lower()
    car_df["cylinders"] = car_df["cylinders"].astype(str).str.strip().str.lower()
    car_df["fuel"] = car_df["fuel"].astype(str).str.strip().str.lower()
    car_df["title_status"] = car_df["title_status"].astype(str).str.strip().str.lower()
    car_df["transmission"] = car_df["transmission"].astype(str).str.strip().str.lower()
    car_df["drive"] = car_df["drive"].astype(str).str.strip().str.lower()
    car_df["type"] = car_df["type"].astype(str).str.strip().str.lower()
    car_df["state"] = car_df["state"].astype(str).str.strip().str.lower()

    car_df["vehicle_age"] = 2026 - car_df["year"]

    car_df["mileage_risk"] = car_df["odometer"].apply(mileage_risk_category)
    car_df["age_risk"] = car_df["vehicle_age"].apply(age_risk_category)
    car_df["title_risk"] = car_df["title_status"].apply(title_risk_category)

    car_df["expected_price"] = best_price_model.predict(car_df[feature_cols])
    car_df["expected_price"] = np.maximum(car_df["expected_price"], 0)

    car_df["price_gap"] = car_df["price"] - car_df["expected_price"]

    car_df["price_gap_percent"] = (
        car_df["price_gap"] / car_df["expected_price"].replace(0, np.nan)
    ) * 100

    car_df["price_gap_percent"] = car_df["price_gap_percent"].replace([np.inf, -np.inf], np.nan)
    car_df["price_gap_percent"] = car_df["price_gap_percent"].fillna(0)

    prediction = risk_classifier_clean.predict(car_df[classification_features_no_leakage])[0]

    probabilities = risk_classifier_clean.predict_proba(car_df[classification_features_no_leakage])
    class_labels = risk_classifier_clean.classes_

    smart_buy_index = list(class_labels).index("smart buy")
    smart_buy_probability = probabilities[:, smart_buy_index][0]

    final_prediction = prediction

    if final_prediction == "smart buy" and smart_buy_probability < 0.75:
        final_prediction = "fair deal"

    result = pd.DataFrame([{
        "actual_price": round(car_df.loc[0, "price"], 2),
        "expected_price": round(car_df.loc[0, "expected_price"], 2),
        "price_gap": round(car_df.loc[0, "price_gap"], 2),
        "price_gap_percent": round(car_df.loc[0, "price_gap_percent"], 2),
        "vehicle_age": car_df.loc[0, "vehicle_age"],
        "mileage_risk": car_df.loc[0, "mileage_risk"],
        "age_risk": car_df.loc[0, "age_risk"],
        "title_risk": car_df.loc[0, "title_risk"],
        "initial_model_prediction": prediction,
        "smart_buy_probability": round(smart_buy_probability, 3),
        "final_buyer_risk_category": final_prediction
    }])

    return result
```
